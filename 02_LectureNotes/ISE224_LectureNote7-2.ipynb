### Getting Started With Pyplot

`Pyplot` is a module of the `matplotlib` library that offers straightforward methods to incorporate various graphical elements, including *lines, images, text, and more, onto the axes* of the existing figure.

````python
# Let's begin our tutorial with a simple graph that uses fundamental Matplotlib code

import matplotlib.pyplot as plt

# plot scatter plot using pandas data object
# Note that the first array appears on the x-axis, 
# and the second array appears on the y-axis of the plot.
plt.plot([1, 2, 3, 4, 5], [4, 3, 7, 6, 8])

# display plot
plt.show()
````

#### add title to figure, and add labels to x-axis and -axis

```
- plt.title(str): add title to the figure ploted  
- plt.xlabel(str): add a label to the x-axis     
- plt.ylabel(str): add a label to the y-axis  
```

````python
# import matplotlib.pyplot as plt
# we already import in the last section, we don't need to import it here again
x = [1, 2, 3, 4, 5]
y = [4, 3, 7, 6, 8]
plt.plot(x, y)

# display plot
plt.show()
````

#### Change color, marker, and line style

```
# color:
plt.plot(x,y,color="assigned color")              # change the color of the line

# line styles: 
plt.plot(x,y,linestype="assigned line style")     # linestype = 'solid','dotted','dashdot','dashed'

# marker:
plt.plot(x,y, marker="assigned marker style", markersize=10, markeredgecolor = "assigned color")      
              # marker: ' 

```
References: 
- [linestyle](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)  
- [marker](https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html)

````python
import matplotlib.pyplot as plt
x = [100,81,64,49,36,25,16,9,4,1]
y = [i**1 for i in x]
y2 = [i**2 for i in x]
y3 = [0.8*i**2 for i in x]
y4 = [0.5*i**2 for i in x]

plt.plot(x, y, color='r', linestyle='solid', marker='.')
plt.plot(x, y2, color='g', linestyle='dotted', marker='o')
plt.plot(x, y3, color='b', linestyle='dashdot', marker='v')
plt.plot(x, y4, color='k', linestyle='dashed', marker='s')

plt.title('Figure Title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
````

#### save figure to 'png' file
To save a plot as a PNG file, you can use the `savefig()` method in Matplotlib. Here's an example code snippet:

```
plt.gcf() # get the current figure object
figure_obj.set_size_inches(width, height) # Change the size of saved file
savefig(fileName) # Save the figure object to fileName
```

````python
x = [1, 2, 3, 4, 5]
y = [i**3 for i in x]
fig = plt.figure()
plt.plot(x, y)
plt.title('Figure Title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
my_plot=plt.gcf() # get the current figure object
my_plot.set_size_inches(6,4)
# Display or save the modified plot
plt.show()  # Display the plot
my_plot.savefig('my_plot.png', dpi=100)  # Save the plot as a PNG file with 100 DPI resolution
````

#### Matplotlib Subplots

The `subplot()` method in Matplotlib allows you to include multiple plots within a single figure. Its syntax is as follows: `plt.subplots(nrows, ncols, index)`, where the three integer arguments represent the desired number of rows and columns, as well as the position of the subplot within the grid.

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 4]

x1 = [100,81,64,49,36,25,16,9,4,1]
y1 = [i**(1/2) for i in x1]

x2 = [1, 2, 3, 4, 5]
y2 = [i**3 for i in x2]


plt.subplot(1,2,1)
plt.plot(x1,y1, color="red", linestyle = "dotted", marker='*', markeredgecolor='k')
plt.title("1st fig.")
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(1,2,2)
plt.plot(x2,y2, color="green",  linestyle = "dashed", marker='p', markeredgecolor='b')
plt.title("2nd fig.")
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
````

### Scatter plot

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [6, 4]

x1 = [36,25,16,9,4,1]
y1 = [i**(1/2) for i in x1]

x2 = [1, 2, 3, 4, 5]
y2 = [i**3 for i in x2]

fig=plt.figure()
plt.plot(x1, y1, linestyle=" ", marker="v", color = 'r')
plt.plot(x2, y2, linestyle=" ", marker="o", color = 'b')
plt.title("Scatter plot - Method: 1")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
````

##### Method 2:  Using `add_axes()` object

An **axes object** is added to figure by the `add_axes()` method. 

The syntax for the axes object is: 
```
ax = fig.add_axes([left,right,width,height]).
```
Then, the scatter plot can be created by:
```
ax.scatter(x, y, color, ...)
```
We can use **ax.set_xlabel**, **ax.set_ylabel**, and **ax.set_title** to add the axis labels and title of the figure.
```
ax.set_xlabel(str)
ax.set_ylabel(str)
ax.set_title(str)
```

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [6, 4]

x1 = [36,25,16,9,4,1]
y1 = [i**(1/2) for i in x1]

x2 = [1, 2, 3, 4, 5]
y2 = [i**3 for i in x2]

fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='b')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('scatter plot: method 2')
plt.show()
````

### Bar chart

A  bar graph presents data with heights and lengths proportional to the values they present.

An axes object is added to figure by the `add_axes()` method. The syntax for the axes object is: 
```
ax = fig.add_axes([left,right,width,height]).
```

With the following code, we can generate the bar chart:
Syntax: 
```
ax.bar(x, height, width, bottom, align)
```

The axes label is a region of the image with data space. 

````python
# Example. Bar chart
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
section = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E', 'Div-F']
students = [25, 19, 33, 28, 14, 21]
ax.bar(section, students)
ax.set_ylabel('People')
plt.show()
````

### Histogram

A histogram is used to understand the distribution of a continuous numerical variable.

Syntax: 
```
ax.hist(x, bins = 'auto')
```

````python
import matplotlib.pyplot as plt
import pandas as pd

# import data
train = pd.read_csv('train.csv')
x = train['Age']

# Add figure object
fig = plt.figure()
# Add add_axes object
ax = fig.add_axes([0, 0, 1, 1])
# plot histogram
ax.hist(x, bins = 'auto')
ax.set_title("Histogram")
ax.set_xlabel("Age")
ax.set_ylabel("Frequency")
plt.show()
````
